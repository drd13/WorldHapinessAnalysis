```python
import numpy as np

def SearchbyKw(keywords,state):  
    pytrends.build_payload(keywords, cat=0, timeframe='2015-01-01 2018-05-25', geo=state, gprop='')
    data = pytrends.interest_over_time()
    TotalSearches={}
    for kw in keywords:
        TotalSearches[kw] = np.sum(data[kw])
    return TotalSearches # returns dictionary of total searches for each keyword
```

```python
SearchbyKw(['suicide'], 'US-TX')
```

```python
def plot_usa_heatmap(state_list,z_list):
    scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
                [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]
    data = [ dict(
            type='choropleth',
            colorscale = scl,
            autocolorscale = False,
            locations = state_list,
            z = z_list,
            locationmode = 'USA-states',
            marker = dict(
                line = dict (
                    color = 'rgb(255,255,255)',
                    width = 2
                )
            ),
            colorbar = dict(
                title = "Visualization"
            )
        ) ]

    layout = dict(
            title = 'Visualization ',
            geo = dict(
                scope='usa',
                projection=dict( type='albers usa' ),
                showlakes = True,
                lakecolor = 'rgb(255, 255, 255)',
            ),
        )
    fig = dict( data=data, layout=layout )
    url = py.plot( fig, filename='d3-cloropleth-map' )
```

```python
abrev = []
z = []
for state in deaths_per_state:
    abrev.append(us_state_abbrev[state])
    z.append(deaths_per_state[state]["Diabetes"]/deaths_per_state[state]["All Causes"])
    print(us_state_abbrev[state])
plot_usa_heatmap(abrev,z)
```
